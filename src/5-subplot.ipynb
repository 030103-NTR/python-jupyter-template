### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange'] 
fig, axs = plt.subplots(2, 2, figsize=(12, 10)) 
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', linewidth=2)
axs[0, 0].set_title('Line Plot', fontsize=14)
axs[0, 0].set_xlabel('X Axis')
axs[0, 0].set_ylabel('Y Axis')
axs[0, 1].bar(categories, values, color=['cyan', 'magenta', 'yellow', 'purple'], edgecolor='black')
axs[0, 1].set_title('Bar Plot', fontsize=14)
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[1, 0].scatter(x, y, color='g', marker='s', s=100, edgecolors='black')
axs[1, 0].set_title('Scatter Plot', fontsize=14)
axs[1, 0].set_xlabel('X Axis')
axs[1, 0].set_ylabel('Y Axis')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart', fontsize=14)
plt.tight_layout()
plt.show()

```
